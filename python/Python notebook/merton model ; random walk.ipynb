```python
# Merton model의 주식수익률 분포
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def merton_random(V0,T,mu,sigma):
    return V0*exp((mu-sigma/2)*T+sqrt(sigma*T)*np.random.normal())
```

```python
zt=0
col=['red','blue','orange','green','violet','pink','yellow','skyblue','gold','silver']
for j in range(10):
    for i in range(255):
        zt=zt+np.random.normal()
        plt.scatter(i,zt,color=col[j])
plt.show()
```

```python
def blackscholes(V,F,r,T,sigma):
    d1=(log(V/F)+(r+pow(sigma,2)/2)*T/(sigma*sqrt(T)))
    d2=d1-sigma*sqrt(T)
    VE=norm(0,1).pdf(d1)*V-F*exp(-r*T)*norm(0,1).pdf(d2)
    sigmaE=norm(0,1).pdf(d1)*sigma*V/VE
    return [VE,sigmaE]
```
