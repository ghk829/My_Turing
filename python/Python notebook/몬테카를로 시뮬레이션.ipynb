```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import *
import numpy as np
import random
```

```python
# 원주율 계산
x=[];y=[];z=[]
for i in range(100000):
    rn1,rn2=random.random(),random.random()
    x.append(rn1)
    y.append(rn2)
    if (pow(rn1,2)+pow(rn2,2)) <1:
        z.append(1)
    else: 
        z.append(0)
print float(sum(z))/float(len(z))*4
col=[]
for i in range(len(z)):
    if z[i]==1:
        col.append('blue')
    else: col.append('red')
plt.scatter(x,y,color=col)
plt.show()
```

```python
# 정규분포 integral 계산
z=[]
for i in range(1000000):
    if 0<np.random.normal() <1.96:
        z.append(1)
    else: z.append(0)
float(sum(z))/float(len(z))
```

```python
x=[];y=[];z=[]
def normal_pdf(x):
    return exp(-1/2*pow(x,2)/sqrt(2*pi))
for i in range(1000):
    rn=np.random.normal()
    x.append(rn)
    y.append(normal_pdf(rn))
    if rn<1.96:
        z.append(1)
    else:
        z.append(0)
col=[]
for i in range(len(z)):
    if z[i]==1:
        col.append('blue')
    else : col.append('red')
print float(sum(z))/float(len(z))
plt.bar(x,y,color=col,width=0.1)
plt.show()
```

```python
# historical experience
z=[]
for i in range(10000):
    rn=random.randint(1,10)
    if rn <=5:
        z.append(0)
    elif rn<=8:
        z.append(1)
    else: z.append(2)
x1=float(z.count(0))/float(len(z))
x2=float(z.count(1))/float(len(z))
x3=1-x1-x2
print x1,x2,x3
```

```python
# m개의 Multivariate Normal
import numpy as np
mean = (1, 2,3);cov = [[1, 0.2,0.3], [0.2,1, 0.7],[0.3,0.7,1]]

z=[]
for i in range(10000):
    x = np.random.multivariate_normal(mean, cov)
    if x[0] < 1 and 1.5<x[1]<2.6: z.append(1)
    else: z.append(0)
print float(sum(z))/float(len(z))
```
