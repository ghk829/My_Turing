```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def int_str(x):
    return str(round(x,3)*100)
```

```python
data1=pd.read_csv('data2.csv')
```

```python
YN=[]
for i in range(data1.shape[0]):
    if data1.ix[i,2] ==9:
        YN.append(0)
    else: YN.append(1)
data1['YN']=YN
```

```python
%matplotlib inline
unique=tuple(data1.ix[:,0].unique())
col=['red','green','blue','yellow','silver','gold']
z=range(21)
for i in range(5):
    plt.figure(i)
    x=[]
    y=[]
    data2=data1.query('cop_id=="%s"'%unique[i])
    for j in range(max(data2.ix[:,4])+1):
        x.append(np.median(data2.query('GRP_ID==%d'%(j)).ix[:,1]))
        y.append(len(data2.query('YN==1 & GRP_ID==%d'%(j)).ix[:,4])/float(len(data2.query('GRP_ID==%d'%(j)).ix[:,4])))
    string=pd.Series(x).apply(int_str)
    plt.xticks(z,string)
    plt.rc('xtick', labelsize=6)
    plt.plot(z,y,color=col[i])
    plt.title('firm'+str(i+1))
    plt.savefig('firm'+str(i+1)+'.png')
```

```python
SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SIZE)                # controls default text sizes
plt.rc('axes', titlesize=SIZE)           # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)          # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plotdata=pd.DataFrame({'x':x,'y':y,'g':z})
```

```python
plotdata.to_csv('plotdata.csv')
```
