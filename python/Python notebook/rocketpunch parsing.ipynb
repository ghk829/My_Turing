```python
import urllib
from bs4 import BeautifulSoup
```

```python
sets=[]
i=1
while True:
    r=urllib.urlopen('https://www.rocketpunch.com/jobs?page=%d'%i)
    soup=BeautifulSoup(r)
    results1=soup.findAll('h4',attrs={'class':'p nowrap jobtitle'})
    for j in range(len(results1)):
        a=results1[j]
        results=a.get_text().split('\n')
        strings=[]
        for k in range(len(results)):
            tmp=results[k].strip()
            if tmp==a.find('small').get_text():
                break
            else : strings.append(tmp)
        sets.append(''.join(strings))
    i+=1
```

```python
sets=set(sets)
sets=list(sets)
```

```python
for i in range(len(sets)):
    print sets[i]
```
