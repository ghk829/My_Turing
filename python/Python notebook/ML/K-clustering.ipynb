```python
from sklearn.datasets import make_blobs
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
import random
X, y= make_blobs(n_samples=150, n_features=2, centers=3,cluster_std=0.5,shuffle=True,random_state=0)
%matplotlib inline
```

```python
def sim(X1,X2):
    a=X1-X2
    dist=sum(a**2)
    return dist
```

```python
def kmeans(X,number,c):
    fset =[X[random.randint(1,X.shape[0]),] for i in range(number)]
    fset=array(fset)
    for l in range(c):
        cluset = [[] for i in range(number)]
        index={}
        for j in range(X.shape[0]):
            pre=[]
            for i in range(fset.shape[0]):
                d=sim(X[j,],fset[i])
                pre.append(d)
            cluset[pre.index(np.min(pre))].append(X[j,])    
            index[j]=pre.index(np.min(pre))    
        cluset=array(cluset)
        for n in range(cluset.shape[0]):
            fset[i]=np.mean(cluset[i],axis=0)
        cluset=cluset.tolist()
    
    return [index,cluset]    
```

```python
result=kmeans(X,3,10)
```

```python
plt.scatter(X[:,0],X[:,1],c=result[0].values());plt.show()
```
