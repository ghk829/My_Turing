```python
from sklearn.datasets import make_blobs
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.cross_validation import train_test_split
X, yy= make_blobs(n_samples=300, n_features=2, centers=2,cluster_std=0.5,shuffle=True,random_state=0)
%matplotlib inline
```

```python
def sim(X1,X2):
    a=X1-X2
    dist=sum(a**2)
    return dist
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,yy,test_size=0.3,random_state=0)
```

```python
def nearclassify(X,ref,refy):
    trainy=[]
    for i in range(len(X)):
        sett=[]
        for j in range(len(ref)):
            sett.append(sim(X[i,],ref[j,]))
        trainy.append(refy[sett.index(np.min(sett))])
    return trainy
```

```python
y=nearclassify(X_test,X_train,y_train)
```

```python
plt.figure(0)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.figure(0)
plt.scatter(X_test[:,0],X_test[:,1],c=y,marker='x')
```
